```python
import os 

diag_path = "API/ParlAI_ex/daily_dialog/train/dialogues_train.txt"
emo_path = "API/ParlAI_ex/daily_dialog/train/dialogues_emotion_train.txt"

end_phrase_token = "__eou__"

def convert_to_int_list(arr):
    return list(map(lambda x: int(x), arr))

with open(emo_path) as f:
    emos = f.readlines()
    emos = list(map(lambda x: convert_to_int_list(x.split(" ")[:-1]), emos)) # drop last "\n"
    # emos = list(map(lambda x: int(x), emos))
    print(emos)

prompts = []
answers = []

with open(diag_path) as f:
    lines = f.readlines()
    for index, line in enumerate(lines):
        diag_list = line.split(end_phrase_token)[:-1]# drop last "\n"

        if len(diag_list) % 2 != 0:
          diag_list = diag_list[:-1]

        emos_item = emos[index]

        # print(len(diag_list[:-1]))
        for j, item in enumerate(diag_list): 

            if j%2==0:
                # print("HERE")
                prompts.append(item)
            else:
                # print("HERE2")
                answers.append(item)


            # print(item, emo_dict[emos_item[j]], j%2==0)

        # print("\n \n \n")
```

```python
import json 

with open('API/ParlAI_ex/friends_dataset_1/emotion-detection-trn.json') as f:
    text = f.read()
    data = json.loads(text)



prompts = []
answers = []

for key, value in data.items():
  
    for v_item in value:
        if "scenes" in v_item:
          scenes = v_item["scenes"]
          for scene in scenes:
            uniq_speakers = []
            utterances = scene["utterances"]
            
            temp_utterances = []
            for utter in utterances:
                speakers = utter["speakers"][0]

                temp_utterances.append(utter)

                if speakers not in uniq_speakers:
                  uniq_speakers.append(speakers)

            if len(uniq_speakers) == 2 and len(temp_utterances) % 2 == 0:
              for j, t in enumerate(temp_utterances):
                  if j%2==0:
                      # print("HERE")
                      prompts.append(t["transcript"])
                  else:
                      # print("HERE2")
                      answers.append(t["transcript"])

                  print(t)

              print(len(temp_utterances))

            print("\n\n")
```
